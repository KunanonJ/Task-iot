<ul>
<li />Verify the design by generating a $\pm$ 0.5 volt sinusoid wave input. The output should swing within 0 - 3 volts.
Plot both input and output waveforms.
</ul>

```python
Vb = -1.5
R1 = 1000 #not equal zero
R2 = (5*R1)/4
R3 = 0
R4 = 1000 #not equal zero

t_vec = np.arange(0,1,0.01)
omega = 10
V_in = 0.5*np.sin(omega*t_vec)
V_out = ((R4/(R3+R4))*(1+R2/R1)*V_in)-(R2/R1)*Vb
```

```python
plt.figure()
plt.plot(t_vec, V_in, 'b-')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')

plt.figure()
plt.plot(t_vec, V_out, 'r-')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
```
