P3. From figure 1, let $V_S(t) = z \cos(y*t)$ volt, $R_S = 1000*x \,\Omega$, $R_L = 500*y \,\Omega$, write a code to plot
$V_L(t)$ compared to $V_S(t)$ in the same graph, for t = 0 - 1 sec

#### Solution

```python
# The values of Rs, RL are defined in the cell above, so you don't have to redefine them 

tvec = np.arange(0,1,0.01)
Vs = Z*np.cos(Y*tvec)
Vl = (Rs/(Rs+Rl))*Vs
plt.figure()
plt.plot(tvec,Vs,'b-',tvec,Vl,'r-')
plt.xlabel('time (s)')
plt.ylabel('volts')
plt.grid(True)
plt.legend(["Vs(t)","Vl(t)"])
```
